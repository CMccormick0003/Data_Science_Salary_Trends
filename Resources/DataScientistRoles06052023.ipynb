```python
import pandas as pd 
from pathlib import Path

df = pd.read_csv('./input/ds_salaries.csv')


#Create path to output clean data to .csv
DSRoles_US_2023_Output = Path('./output/DSRoles_US_2023.csv')
DSRoles_US_2023_Output.parent.mkdir(parents=True, exist_ok=True) 

#print('Size of Data Scientist Salaries frame is :',df.shape)
df
```

```python
#Refine Data to 2023 work year
df_refined2023 = df[(df["work_year"]==2023)]
df_refined2023
```

```python
#Refine employee_residence data to US
df_refined2023_USOnly = df_refined2023[(df_refined2023["employee_residence"]== "US")]
df_refined2023_USOnly 
```

```python
#Create Columns for codes
df_refined2023_USOnly["employment_type_code"]=" "
df_refined2023_USOnly["job category code"]=" "
df_refined2023_USOnly["salary_in_usd_code"]=" "
df_refined2023_USOnly["company_size_code"]=" "
df_refined2023_USOnly["experience_level_code"]=" "
```

```python
#Update employment_type_code
df_refined2023_USOnly.loc[df_refined2023_USOnly['employment_type'] == 'CT','employment_type_code']='1'  
df_refined2023_USOnly.loc[df_refined2023_USOnly['employment_type'] == 'FT','employment_type_code']='2' 
#Update company_size
df_refined2023_USOnly.loc[df_refined2023_USOnly['company_size'] == 'S','company_size_code']='1' 
df_refined2023_USOnly.loc[df_refined2023_USOnly['company_size'] == 'M','company_size_code']='2'
df_refined2023_USOnly.loc[df_refined2023_USOnly['company_size'] == 'L','company_size_code']='3'
#Update experience_level_code
df_refined2023_USOnly.loc[df_refined2023_USOnly['experience_level'] == 'EN','experience_level_code']='1'
df_refined2023_USOnly.loc[df_refined2023_USOnly['experience_level'] == 'MI','experience_level_code']='2'
df_refined2023_USOnly.loc[df_refined2023_USOnly['experience_level'] == 'SE','experience_level_code']='3'
```

```python
#Update job category code
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Analyst'),'job category code']='1' 
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Architect'),'job category code']='2'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Engineer'),'job category code']='3' 
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Data Manager'),'job category code']='4' 
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Data Scientist'),'job category code']='5'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Data Science'),'job category code']='5'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Machine Learning Scientist'),'job category code']='5'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Developer'),'job category code']='6'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Data Specialist'),'job category code']='7'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Head of Data'),'job category code']='7'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Data Lead'),'job category code']='7'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Applied Scientist'),'job category code']='8'
df_refined2023_USOnly.loc[df_refined2023_USOnly['job_title'].str.contains('Research Scientist'),'job category code']='8'
```

```python
#Update salary_in_usd_code
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(0, 50000, 'both'),'salary_in_usd_code']='1'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(51000, 100000, 'both'),'salary_in_usd_code']='2'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(101000, 150000, 'both'),'salary_in_usd_code']='3'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(151000, 200000, 'both'),'salary_in_usd_code']='4'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(201000, 250000, 'both'),'salary_in_usd_code']='5'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(251000, 300000, 'both'),'salary_in_usd_code']='6'
df_refined2023_USOnly.loc[df_refined2023_USOnly['salary_in_usd'].between(301000, 350000, 'both'),'salary_in_usd_code']='7'
```

```python
df_refined_2023_US = df_refined2023_USOnly[["employment_type_code","experience_level_code","job category code","salary_in_usd_code","company_size_code","remote_ratio"]]
```
