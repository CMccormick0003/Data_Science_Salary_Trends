```python
# import data
df = pd.read_csv("ds_salaries.csv")
df.head()
```

```python
# all column name
df.columns
```

```python
df = df.loc[df["job_title"].isin(['Data Scientist', 'Data Engineer','Data Analyst'])]
df
```

```python
df = df.loc[df["employment_type"].isin(['FT', 'CT','PT'])]
df
```

```python
df = df.loc[df["work_year"] == 2023]
df
```

```python
df.drop('salary', inplace=True, axis=1)
df
```

```python
df.drop('salary_currency', inplace=True, axis=1)
```

```python
df
```

```python
df.drop('employee_residence', inplace=True, axis=1)
df
```

```python
df = df.loc[df["company_location"].isin(['US'])]
df
```

```python
X = df[["experience_level","employment_type", "job_title", "remote_ratio", "company_size"]]
y = df["salary_in_usd"]
```

```python
X
```

```python
dummies = pd.get_dummies(X, columns = ['experience_level', 'employment_type', 'job_title', 'company_size'], drop_first = False)
```

```python
X=X.drop(['experience_level', 'employment_type', 'job_title', 'company_size'], axis = 1)
```

```python
X = pd.concat([X, dummies], axis = 1)
```

```python
X
```

```python
# split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
# instantiate 
model = LinearRegression()
# fit
model.fit(X_train, y_train)
# predict
y_pred = model.predict(X_test)
```

```python
# model evaluation
score = r2_score(y_test, y_pred)
print(score)
```
